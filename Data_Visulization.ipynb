**Import Data**

```python
import pandas as pd
import matplotlib.pylab as plt

df = pd.read_csv('data/lab2_Topic_Survey_Assignment.csv')
df.head()
```

**Clean data**

```python
df['Total'] = df.loc[:,'Very interested':'Not interested'].sum(axis = 1)
df = df.rename(columns = {'Unnamed: 0':'Data Area'})
df
```

**Use list to contain slicing data values**

```python
vi = df['Very interested']
si = df['Somewhat interested']
ni = df['Not interested']
```

**Calculate the percentage**

```python
vi_per = df['Very interested']/df['Total']*100
vi_per = [round(x, 2) for x in vi_per]
si_per = df['Somewhat interested']/df['Total']*100
si_per = [round(x, 2) for x in si_per]
ni_per = df['Not interested']/df['Total']*100
ni_per = [round(x, 2) for x in ni_per]

new_df = df.drop(df.iloc[:,1:5], axis=1)
new_df['Very interested'] = vi_per
new_df['Somewhat interested'] = si_per
new_df['Not interested'] = ni_per
new_df = new_df.sort_values(by='Very interested', ascending=False)

# print the sorted DataFrame
new_df 
```

```python
x_pos = [i for i in range(len(vi_per))]
bar_width = 0.25
fig, ax = plt.subplots(figsize = (15,6))

ax.bar(x_pos, new_df['Very interested'], bar_width, color='#00b300', label='Very Interested')
for i, v in enumerate(new_df['Very interested']):
    ax.text(i - 0.1, v + 1, str(v), color='black', fontsize=8)

ax.bar([i + bar_width for i in x_pos], new_df['Somewhat interested'], bar_width, color='#4dc3ff', label='Somewhat Interested')
for i, v in enumerate(new_df['Somewhat interested']):
    ax.text(i + bar_width -0.1, v + 1, str(v), color='black', fontsize=8)

ax.bar([i + 2*bar_width for i in x_pos], new_df['Not interested'], bar_width, color='#cc3300', label='Not Interested')
for i, v in enumerate(new_df['Not interested']):
    ax.text(i + 2*bar_width - 0.1, v + 1, str(v), color='black', fontsize=8)

# remove y
plt.gca().yaxis.set_visible(False)

# remove border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

ax.set_xticks([i + bar_width for i in x_pos])
ax.set_xticklabels(new_df['Data Area'])
ax.set_title("Percentage of Respondents's Interest In Data Science Areas")
ax.legend()

plt.show()
```

# Solve by another way 

```python
import pandas as pd
import matplotlib.pylab as plt

df = pd.read_csv('lab2_Topic_Survey_Assignment.csv')
df['Total'] = df.loc[:,'Very interested':'Not interested'].sum(axis = 1)
df = df.rename(columns = {'Unnamed: 0':'Data Area'})
df
```

**Use 2 lines code to take percentage of data instead of list**

```python
df[['Very interested','Somewhat interested','Not interested']] = df[['Very interested','Somewhat interested','Not interested']].apply(lambda x: ((x/df['Total'])*100).astype('float'))
df= df.round(decimals=2).sort_values(by = 'Very interested', ascending = False)
df
```

```python
x_pos = [i for i in range(len(df['Data Area']))]
bar_width = 0.25
fig, ax = plt.subplots(figsize = (15,6))

ax.bar(x_pos, df['Very interested'], bar_width, color='#00b300', label='Very Interested')
for i, v in enumerate(df['Very interested']):
    ax.text(i - 0.1, v + 1, str(v), color='black', fontsize=8)

ax.bar([i + bar_width for i in x_pos], df['Somewhat interested'], bar_width, color='#4dc3ff', label='Somewhat Interested')
for i, v in enumerate(df['Somewhat interested']):
    ax.text(i + bar_width -0.1, v + 1, str(v), color='black', fontsize=8)

ax.bar([i + 2*bar_width for i in x_pos], df['Not interested'], bar_width, color='#cc3300', label='Not Interested')
for i, v in enumerate(df['Not interested']):
    ax.text(i + 2*bar_width - 0.07, v + 1, str(v), color='black', fontsize=8)

# Remove the x-axis only
plt.gca().yaxis.set_visible(False)

# remove border
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.gca().spines['left'].set_visible(False)

ax.set_title("Percentage of Respondents's Interest In Data Science Area")   
ax.set_xticks([i + bar_width for i in x_pos])
ax.set_xticklabels(df['Data Area'])
ax.legend()

plt.show()
```
