# **Data Preprocessing**

# **1. Handling missing values in columns**

Mode

Mean

Median

```python
question_cols = df.columns[df.isin(['?']).any()]
question_cols
```

```python
import numpy as np
df =  df.replace("?",np.nan)
question_cols = df.columns[df.isin(['?']).any()]
question_cols
```

```python
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])
df['bore'] = pd.to_numeric(df['bore'])
df['stroke'] = pd.to_numeric(df['stroke'])
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'])
df['price'] = pd.to_numeric(df['price'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

sns.histplot(data=df, x="normalized-losses", kde=True, ax=ax1)
ax1.set_title("Before Filling Missing Values")

mean_normal = df['normalized-losses'].median(skipna=True)
df['normalized-losses'].fillna(mean_normal, inplace=True)

sns.histplot(data=df, x="normalized-losses", kde=True, ax=ax2)
ax2.set_title("After Filling Missing Values")

plt.show()
```

```python
mean_peak = df['peak-rpm'].mean(skipna=True)
mean_bore = df['bore'].mean(skipna=True)
mean_stroke = df['stroke'].mean(skipna=True)
mean_hp = df['horsepower'].mean(skipna=True)
mean_price = df['price'].mean(skipna=True)
mode_doors = df['num-of-doors'].mode()

# fill the NaN values with the mean of the column
df['peak-rpm'] = df['peak-rpm'].fillna(mean_peak)
df['bore'] = df['bore'].fillna(mean_bore)
df['stroke'] = df['stroke'].fillna(mean_stroke)
df['horsepower'] = df['horsepower'].fillna(mean_hp)
df['price'] = df['price'].fillna(mean_price)
df['num-of-doors'] = df['num-of-doors'].fillna(mode_doors)

df.info()
```

# **Binning**

```python
bins_price = np.linspace(min(df['price']), max(df['price']),4)
group_names = ['Low', 'Medium','High']
df['price-binned']= pd.cut(df['price'],bins_price, labels = group_names, include_lowest= True)
df[['price','price-binned']]
```

# **Introduction to Matplotlib**

```python
brand_car = df.groupby('body-style')['body-style'].count()
name = brand_car.index
value = brand_car.values

brand_car
```

**Line Chart**

```python
import matplotlib.pyplot as plt
figure = plt.figure(figsize = (8,4))
plt.plot(name,value)
plt.xlabel('name', fontsize = 15, fontweight = 'bold', c = 'red')
plt.ylabel('value')
plt.title('Number of car use in each brand')
plt.show()
```

**Bar chart**

```python
figure = plt.figure(figsize = (8,4))
plt.bar(name,value, color = 'red', width=0.5)
plt.xlabel('name', fontsize = 15)
plt.ylabel('value')
plt.title('Number of car use in each brand')
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10,5))

ax1.bar(name,value, color = 'red', width=0.5)
ax1.set_xlabel('name')
ax1.set_ylabel('value')
ax1.set_title('Number of car use in each brand')

ax2.plot(name,value, color = 'blue')
ax2.set_xlabel('name')
ax2.set_ylabel('value')
ax2.set_title('Number of car use in each brand')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.pie(value, labels=name, autopct='%.1f%%')
ax.set_title('Number of car use in each brand')
ax.legend(loc = 'upper right',bbox_to_anchor=(1.5, 1))
```
