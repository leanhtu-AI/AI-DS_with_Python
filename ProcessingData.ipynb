# **Data Wrangling**

**Đổi tên cột**

```python
headers = ["symbolling","normalized-losses","make","fuel-type","aspiration","num-of-doors","body-style", "drive-wheels","engine-location","wheel-base","length","width","height","curb-weight","engine-type", "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower","peak-rpm","city-mpg","highway-mpg","price"]
df.columns=headers
df
```

**Thay thế ? = mean**

```python
# find all rows with '?' values in any column
cols_with_question_marks = df.columns[df.isin(['?']).any()]

cols_with_question_marks
```

```python
# replace '?' values with NaN
df = df.replace('?', np.nan)

# convert the column to numeric type
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])
df['bore'] = pd.to_numeric(df['bore'])
df['stroke'] = pd.to_numeric(df['stroke'])
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'])
df['price'] = pd.to_numeric(df['price'])

# calculate the mean of the column (ignoring NaN values)
mean1 = df['normalized-losses'].mean(skipna=True)
mean2 = df['peak-rpm'].mean(skipna=True)
mean3 = df['bore'].mean(skipna=True)
mean4 = df['stroke'].mean(skipna=True)
mean5 = df['horsepower'].mean(skipna=True)
mean6 = df['price'].mean(skipna=True)
mode_doors = df['num-of-doors'].mode()

# fill the NaN values with the mean of the column
df['normalized-losses'] = df['normalized-losses'].fillna(mean1)
df['peak-rpm'] = df['peak-rpm'].fillna(mean2)
df['bore'] = df['bore'].fillna(mean3)
df['stroke'] = df['stroke'].fillna(mean4)
df['horsepower'] = df['horsepower'].fillna(mean5)
df['price'] = df['price'].fillna(mean6)
df['num-of-doors'] = df['num-of-doors'].fillna(mode_doors)

```

# **Data Normalizing 3 columns**

**Simple feature scaling**

```python
#take a dataframe from df
df1 = df

#x_new = x_old / x_max 
df1['length'] = df1['length']/df1['length'].max()
df1['height'] = df1['height']/df1['height'].max()
df1['width'] = df1['width']/df1['width'].max()

#print
df1[['length','height','width']].head()
```

**Min - Max**

```python
df2 = df

#x_new = x_old - x_min / x_max - x_min
df2['length'] = (df2['length'] - df2['length'].min())/ (df2['length'].max() - df2['length'].min())
df2['height'] = (df2['height'] - df2['height'].min())/ (df2['height'].max() - df2['height'].min())
df2['width'] = (df2['width'] - df2['width'].min())/ (df2['width'].max() - df2['width'].min())

#print
df2[['length','height','width']].head()
```

**Z-Score**

```python
df3 = df
#x_new = x_old - x_min / x_max - x_min
df3['length'] = (df3['length'] - df3['length'].mean())/ df3['length'].std()
df3['height'] = (df3['height'] - df3['height'].mean())/ df3['height'].std()
df3['width'] = (df3['width'] - df3['width'].mean())/ df3['width'].std()

#print
df3[['length','height','width']].head()
```

# **Binning**

```python
#price
import numpy as np
bins_price = np.linspace(min(df['price']), max(df['price']),4)
group_names = ['Low', 'Medium','High']
df['price-binned']= pd.cut(df['price'],bins_price, labels = group_names, include_lowest = True)
df[['price','price-binned']]
```

```python
#Engine-size
bins_engine = np.linspace(min(df['engine-size']), max(df['engine-size']),4)
group_names = ['Small', 'Normal','Big']
df['engine-binned']= pd.cut(df['engine-size'],bins_engine, labels = group_names, include_lowest = True)
df[['engine-size','engine-binned']]
```

```python
#Horsepower
bins_horsepower = np.linspace(min(df['horsepower']), max(df['horsepower']),5)
group_names = ['Slow', 'Medium','Fast', 'Super Fast']
df['horsepower-binned']= pd.cut(df['horsepower'],bins_horsepower, labels = group_names, include_lowest = True)
df[['horsepower','horsepower-binned']]
```
